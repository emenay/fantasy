```python
import pandas as pd
import espnfantasyfootball as espn 
import espnsecrets
import numpy as np
```

```python
week = 1
```

```python
league = espn.FantasyLeague(league_id=espnsecrets.league_id, year=2023,
                            espn_s2=espnsecrets.espn_s2, swid=espnsecrets.swid)
data = league.get_league_data(week)

data
```

```python
#   Highest scoring player
arr = data["PlayerScoreActual"].to_numpy()
maxScore = max(arr)
maxScoreIndex = np.where(arr == maxScore)[0][0]
maxScorer = data["PlayerName"][maxScoreIndex]

#   Highest scoring team
#   Lowest scoring team
team_num = data["PlayerFantasyTeam"].to_numpy()
team_names = np.unique(data["TeamName"].to_numpy())
team_score = {}
for i in team_names:
    team_score[i] = (np.where(np.logical_and(data["TeamName"]==i, data["PlayerRosterSlot"] != 'Bench'), data["PlayerScoreActual"], 0).sum())

#   Most surprising player
x= list((data["PlayerScoreProjected"]-data["PlayerScoreActual"]).abs())
index = x.index(max(x))
string1, string2 = "", ""
if data["PlayerScoreActual"][index] > data["PlayerScoreProjected"][index]:
    string1 = "shocking"
    string2 = "surpass"
else:
    string1 = "disappointing"
    string2 = "shatter"

#   Most lost potential
bench_score = {}
for i in team_names:
    bench_score[i] = (np.where(np.logical_and(data["TeamName"]==i, data["PlayerRosterSlot"] == 'Bench'), data["PlayerScoreActual"], 0).sum())

potential_team = max(bench_score, key = bench_score.get)
```

```python
def bold_text(text):
    bold_start = '\033[1m'
    bold_end = '\033[0m'
    return bold_start + text + bold_end
```

```python
new_team_names = {}
new_team_names['Alabama Peach Eaters'] = bold_text('Alabama Peach Eaters') 
new_team_names['Blitzburg Busters'] = bold_text('Blitzburg Busters')
new_team_names['California Cucks'] = bold_text('California Cucks')
new_team_names['Fayetteville Fayettevilles '] = bold_text('Fayetteville Fayettevilles')
new_team_names["It's corn 🌽"] = bold_text('Sack City Btch')
new_team_names['Player 1'] = bold_text('Quon Solo')
new_team_names['Richmond Red Necks'] = bold_text('Richmond Red Necks')
new_team_names['Riley Read Option'] = bold_text('Ertz When I Pee')
new_team_names['Ryans Burner'] = bold_text('Ryans Burner')
new_team_names['Team Chheda'] = bold_text('jinny\'s team 🙏🏽')
new_team_names['Team Trocinski'] = bold_text('Cheeseheads')
new_team_names['Team Vanhoy'] = bold_text('Sacks and the city')
new_team_names['Washington  Whippersnappers'] = bold_text('Washinton Whippersnappers')
new_team_names['Watson Sucks'] = bold_text('Hurts my McAnus')
```

```python

print("- {} with a whopping {} points as week {}'s highest scoring player.".format(bold_text(maxScorer), maxScore, week))

print("- Highest scoring team this week: {}".format(bold_text(new_team_names[max(team_score, key = team_score.get)])))
print("- Lowest scoring team this week: {}".format(bold_text(new_team_names[min(team_score, key = team_score.get)])))

print("- In a {} turn of events, {} {} all expectations, scoring {} instead of the projected {}.".format(string1, bold_text(data["PlayerName"][index]), string2, data["PlayerScoreActual"][index], data["PlayerScoreProjected"][index]))

print("- {} lost some potential by leaving {} points on the bench this week.".format(bold_text(new_team_names[potential_team]), bench_score[potential_team]))

print("- Tune in next week for a summary of week {}.".format(week+1))
```
